```python
import requests, json
import pandas as pd
```

```python
source_url = "https://api.um.warszawa.pl/api/action/datastore_search?resource_id=ed6217dd-c8d0-4f7b-8bed-3b7eb81a95ba&limit=100"

response = requests.get(source_url)

jsonData = json.loads(response.text)

data = jsonData['result']['records']

print(data)

with open("out.json","w") as outfile:
    json.dump(data, outfile, indent = True, ensure_ascii=False)
```
